```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
model_results = pd.read_csv('model_results.csv')
for (columnName, columnData) in model_results.iteritems():
    print(f"{columnName} : {mean_squared_error(model_results['2013 Real TEY'],columnData)}")
```

```python
for (columnName, columnData) in model_results.iteritems():
    plt.plot(model_results[f'{columnName}'].loc[3000:3100])
plt.legend(['TEY Actual Values','LinReg','4-I NN','5-I NN','1HL 3N'])
plt.title('Random Sample of TEY Predicted Values')
plt.show()
```

```python
plt.plot(model_results['2013 Real TEY'].loc[2000:2100])
plt.plot(model_results['Regression on PCA'].loc[2000:2100])
plt.plot(model_results['CDP vs TEY Regression'].loc[2000:2100])
plt.plot(model_results['GTEP vs TEY Regression'].loc[2000:2100])
plt.title('Random Sample of TEY Predicted Values ctd.')
plt.legend(['2013 Real TEY','Regresson on PCA','CDP vs. TEY','GTEP vs. TEY'])
plt.show()
```
