```python
# libraries 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
#correlation_mat = df_2013.corr()
df = [pd.read_csv(f'./../../Data/Train/{x.strip()}') for x in os.popen('ls ../../Data/Train/')]
data = pd.concat(df)
correlation_mat = data.corr()
sns.heatmap(correlation_mat, annot = True)

plt.title("Correlation matrix of data All Train Data (2011,12)")
plt.xlabel("Gas Turbine Variables")
plt.ylabel("Gas Turbine Variables")
plt.show()
```

```python
corr_pairs = correlation_mat.unstack()
strong_pairs = corr_pairs[abs(corr_pairs) > 0.7]
strong_pairs = strong_pairs[abs(strong_pairs) != 1]
strong_pairs = strong_pairs.sort_values(ascending=False,key=lambda x: abs(x)) 
print(strong_pairs)
```
